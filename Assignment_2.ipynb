```python
Q10). How can you tell the difference between break and continue?

Solution :-
    break is used to break the flow of the program in a loop.when break is used it breaks the 
    nearest loop. continue is used to continue the program like dont go for further steps in that loop 
    after seeing continue, just jump to next iteration 
    most often both of these are used with if block.
    break is used to stop the iteration in an endless loop but continue cannot do this.
    Example.
    i=0
    while(True):
        print("infinite")
        i=+1
        if (i==9):
            break
    Here it is actually a endless loop but in if condition when value of i = 9 at that time it breaks 
    the while loop.
    
    Example
    i=0
    while i<9:
        i+=1
        if i==3:
            continue
        print(i)
    Here, it will print values from 1 to 9 except value 3 here after continue it again goes to
    while condition.
        
    
```

```python
Q11). In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

Solution :-
    like to say logically there is no difference between range(10), range(0, 10), and range(0, 10, 1)
    range(10) produces number ranging from 0 to 9.
    range(0,10),range(0,10,1) also produces same numbers
    difference is only about arguments it takes.It basically takes 3 aruguments
    first is starting number for our range,second is the end point of the range but that number is 
    not included and third is the step argument.
    range(0,10,1) here start is 0 and end is 9 as 10 is not included and step argument is 1
    In range(0,10) here start is 0 and end is 9 here no step argument mentioned but by default 
    step argument is 1,In range(10) here start argument is not given so by default it takes start as 
    0 then end is 9 and step argument is again by default 1.
    
```

```python
Q12). Write a short program that prints the numbers 1 to 10 using a for loop. 
Then write an equivalent program that prints the numbers 1 to 10 using a while loop.

Solution :-
1) For loop
for i in range(1,11):
    print(i)
    
2) While loop
i=1
while i <= 10:
    print(i)
    i+=1
```
